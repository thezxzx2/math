# Relación de proposiciones abiertas con conjuntos

$p = "x \text{ es un número par menor que 10 }; x \in N$

**Negación**. La negación se relaciona con el complemento de un conjunto.

#### Ejemplo

$$a = "x \text{ es un dígito par}" \hspace{4cm} \sim a = "x \text{ no es un dígito par}"$$

#### Solución
